```python
df_full = pandas.read_csv("./FineFieldScan_gm2ringsim_3.csv", header=None, names=['time', 'r', 'phi', 'x', 'y', 'z', 
                                                                          'field_1', 'field_2', 'field_3', 
                                                                          'field_4', 'field_5', 'field_6', ])
df = df_full#.sample(100000)
df["field_Br"] = np.sqrt(df['field_1']**2 + df["field_3"]**2)
```

```python
interpolations = {}

whichZero =  ['y', 'r', 'y','y', 'r',  'y', 'y']
whichPoint = [ 1,  7113, 1,  1,   7113, 1,   1]
whichPlot =  ['r', 'y', 'r','r', 'y',  'r', 'r']

for i, field_comp in enumerate([1,2,3,4,5,6,"Br"]):
    print(i, field_comp)
    #fig,ax = plt.subplots(figsize=(15,7))

    dfi = df.loc[df[whichZero[i]] == whichPoint[i]]
    #print(dfi)
    
    x = dfi['phi']
    y = dfi[whichPlot[i]]
    rho = dfi['field_'+str(field_comp)]

    # Set up a regular grid of interpolation points
    xi, yi = np.linspace(x.min(), x.max(), 3000), np.linspace(y.min(), y.max(), 3000)
    xi, yi = np.meshgrid(xi, yi)

    # Interpolate; there's also method='cubic' for 2-D data such as here
    zi = scipy.interpolate.griddata((x, y), rho, (xi, yi), method='cubic')
    interpolations[field_comp] = zi
```

```python
fig,ax = plt.subplots(2,3,figsize=(20,10), sharex=True)
ax = flatten2dArray(ax)
field_names = [r"$B_{x}$ [Tesla]", r"$B_{y}$ [Tesla]", r"$B_{z}$ [Tesla]",
               r"$E_{x}$", r"$E_{y}$", r"$E_{z}$"]
for field_comp in range(1,7):
    axi = ax[field_comp-1]
    plt.sca(axi)
    #fig,ax = plt.subplots(figsize=(15,7))

    x = df['phi']
    y = df[whichPlot[field_comp - 1]]
    rho = df['field_'+str(field_comp)]
    plt.title("Field Component: "+field_names[field_comp - 1])
    axi.set_ylabel(whichPlot[field_comp - 1]+" [mm]")
    plt.xlabel(r"$\phi$ [rad]")

    # Set up a regular grid of interpolation points
    xi, yi = np.linspace(x.min(), x.max(), 3000), np.linspace(y.min(), y.max(), 3000)
    xi, yi = np.meshgrid(xi, yi)

    # Interpolate; there's also method='cubic' for 2-D data such as here
    zi = interpolations[field_comp]

    plt.imshow(zi, vmin=zi.min(), vmax=zi.max(), origin='lower',
               extent=[x.min(), x.max(), y.min(), y.max()],
                aspect='auto')


    plt.colorbar()
    plt.grid()

plt.suptitle("Slices in Storage Region", y=1.02, fontsize=18)
plt.tight_layout()
plt.show()
```

### radial field vs. y

```python
interpolations = {}
for i in range(-28,30,2):
    print("y =", i)
    dfi = df.loc[df['y'] == i]
    
    x = dfi['phi']
    y = dfi['r']
    rho = dfi['field_Br']

    # Set up a regular grid of interpolation points
    xi, yi = np.linspace(x.min(), x.max(), 3000), np.linspace(y.min(), y.max(), 3000)
    xi, yi = np.meshgrid(xi, yi)

    # Interpolate; there's also method='cubic' for 2-D data such as here
    zi = scipy.interpolate.griddata((x, y), rho, (xi, yi), method='cubic')
    interpolations[i] = zi
print(len(interpolations))
```

```python
fig,ax = plt.subplots(5,6,figsize=(20,10),sharex=True, sharey=True)
ax = flatten2dArray(ax)

for i, yi in enumerate(interpolations):
    x = df['phi']
    y = df['r']
    plt.sca(ax[i])
    
    plt.imshow(interpolations[yi], 
               #vmin=0.00001, vmax=1, 
               origin='lower',
               extent=[x.min(), x.max(), y.min(), y.max()],
               norm = matplotlib.colors.LogNorm(),
               aspect='auto')
    plt.colorbar()
    plt.title("Slice of y = "+str(yi)+" mm")
plt.tight_layout()
plt.ylim(6850,7050)
plt.show()
```

### average radial field for various slices of y

```python
ywidth = 2
fields = {}
# rmin = 7112-45 #decay points
rmax = 7112+45 #decay points
rmin = 7112-245 #decay points and calos
for i in range(-28,30,ywidth):
    print(i)
    dfi = (df.loc[df['y'] > i].loc[df['y'] <= i+ywidth]
           .loc[df['r'] >= rmin].loc[df['r'] <= rmax])
    #print(dfi)
    phi = list(set(dfi['phi']))
    phi.sort()
    mean_field = []
    for pi in phi:
        dfii = dfi.loc[df['phi'] == pi]
        mean_field.append(dfii['field_Br'].mean())
    fields[i] = mean_field
```

```python
fig,ax = plt.subplots(figsize=(7,5))

#plt.scatter(dfi['phi'], dfi['field_Br'], c=dfi['y'])
plt.plot(phi, mean_field)
plt.title("Mean Radial Field vs. Azimuth in gm2ringsim")
plt.xlabel(r"$\phi$ [rad]")
plt.ylabel(r"$B_{r}$ [T]")
plt.grid()
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(15,10))

#plt.scatter(dfi['phi'], dfi['field_Br'], c=dfi['y'])
for i in fields:
    fi = fields[i]
    plt.plot(phi, fi, label=str(i))
plt.title("Mean Radial Field vs. Azimuth in gm2ringsim")
plt.xlabel(r"$\phi$ [rad]")
plt.ylabel(r"$B_{r}$ [T]")
plt.grid()
plt.legend(ncol=6)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(15,10))

#plt.scatter(dfi['phi'], dfi['field_Br'], c=dfi['y'])
for i in fields:
    fi = fields[i]
    plt.plot(phi, [x- fi[0] for x in fi], label=str(i))
plt.title("Mean Radial Field vs. Azimuth in gm2ringsim")
plt.xlabel(r"$\phi$ [rad]")
plt.ylabel(r"$B_{r}$ [T]")
plt.grid()
plt.legend(ncol=6)
plt.show()
```

```python
fig,ax = plt.subplots(5,6,figsize=(15,10), sharey=True, sharex=True)
ax = flatten2dArray(ax)
#plt.scatter(dfi['phi'], dfi['field_Br'], c=dfi['y'])
for fi, i in enumerate(fields):
    plt.sca(ax[fi])
    fi = fields[i]
    plt.plot(phi, fi, label=str(i))
    plt.title("y = "+str(i))
    plt.grid()
plt.suptitle("Mean Radial Field vs. Azimuth in gm2ringsim", y=1.02)
plt.xlabel(r"$\phi$ [rad]")
plt.ylabel(r"$B_{r}$ [T]")
plt.grid()
plt.tight_layout()
plt.show()
```
