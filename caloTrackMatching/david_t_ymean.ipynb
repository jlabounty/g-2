```python
f = r.TFile("./data/ADT_run2c/results_ADTscan_June9_2_3.root")
```

```python
c = r.TCanvas("c","c",3200,1600)
c.Divide(6,4)
for calo in range(1,25):
    c.cd(calo)
    h = f.Get("clusters/randclusters"+str(calo))
#     h.GetYaxis().SetRangeUser(0,1000)
    h2 = h.Project3D("y").Clone()
    h2.SetTitle("y-Position in Calo "+str(calo)+" vs. Time; Time [randomized, #mus]; y [crystal widths from bottom]")
    h2.DrawCopy("colz")
#     break
c.Draw()
```
