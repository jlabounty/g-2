```python
eLow = 1100 # MeV
eHigh = 2400
```

```python
nbins = hy.GetNbinsX()
ndivs = 10
c = r.TCanvas("c","c",3600,1600)
c.Divide(5,2)
for i in range(ndivs):
    binlow = int(nbins/ndivs)*i + 1
    binhigh = int(nbins/ndivs)*(i+1)
    print(binlow,binhigh)
    
    hy.GetXaxis().SetRange(binlow,binhigh)
    hy.GetZaxis().SetRangeUser(eLow,eHigh)
    
    hy_y = hy.Project3D("y").Clone()
    hy_y.SetTitle("y-Position [Time bins "+str(hy.GetXaxis().GetBinCenter(binlow) - hy.GetXaxis().GetBinWidth(binlow)/2.0)+" - "+str(hy.GetXaxis().GetBinCenter(binhigh)  + hy.GetXaxis().GetBinWidth(binlow)/2.0)+"]")
    
    c.cd(i+1)
    hy_y.DrawCopy()
    r.gPad.SetLogy()

c.Draw()
```

```python
for calo in range(1,25):
    print("Starting calo", calo)
    f = r.TFile("./data/results_westPositionVsTime_EndGame_New.root")
    hy = f.Get("farline").Get("calo"+str(calo)).Get(yHistName).Clone("hy_"+str(calo))
    nbins = hy.GetNbinsX()
    ndivs = 10
    c = r.TCanvas("c","c",3600,1600)
    c.Divide(5,2)
    for i in range(ndivs):
        binlow = int(nbins/ndivs)*i + 1
        binhigh = int(nbins/ndivs)*(i+1)
        #print(binlow,binhigh)

        hy.GetXaxis().SetRange(binlow,binhigh)
        hy.GetZaxis().SetRangeUser(eLow,eHigh)

        hy_y = hy.Project3D("y").Clone()
        hy_y.SetTitle(("Calo "+str(calo)+" [Time range "
                       +str(hy.GetXaxis().GetBinCenter(binlow) - hy.GetXaxis().GetBinWidth(binlow)/2.0)
                       +" - "+str(hy.GetXaxis().GetBinCenter(binhigh)  + hy.GetXaxis().GetBinWidth(binlow)/2.0)
                       +" , Energy range "+str(eLow)+" - "+str(eHigh)+" MeV "
                       +"]"))

        c.cd(i+1)
        hy_y.DrawCopy()
        r.gPad.SetLogy()

    c.Draw()
    c.Print("./images/TimeDependance_West_Calo"+str(calo).zfill(2)+".png")
```

### Now lets make the same plot but for energy bins

```python
tLow = 30
tHigh = 600
for calo in range(1,25):
    print("Starting calo", calo)
    f = r.TFile("./data/results_eastPositionVsTime_EndGame_New.root")
    hy = f.Get("farline").Get("calo"+str(calo)).Get(yHistName).Clone("hy_"+str(calo))
    nbins = hy.GetNbinsZ()
    ndivs = 14
    c = r.TCanvas("c","c",4800,1600)
    c.Divide(7,2)
    for i in range(ndivs):
        binlow = int(nbins/ndivs)*i + 1
        binhigh = int(nbins/ndivs)*(i+1)
        #print(binlow,binhigh)

        hy.GetZaxis().SetRange(binlow,binhigh)
        hy.GetXaxis().SetRangeUser(tLow,tHigh)

        hy_y = hy.Project3D("y").Clone()
        hy_y.SetTitle(("Calo "+str(calo)+" [Energy range "
                       +str(hy.GetZaxis().GetBinCenter(binlow) - hy.GetZaxis().GetBinWidth(binlow)/2.0)
                       +" - "+str(hy.GetZaxis().GetBinCenter(binhigh)  + hy.GetZaxis().GetBinWidth(binlow)/2.0)
                       +" , Time range "+str(tLow)+" - "+str(tHigh)+" MeV "
                       +"]"))

        c.cd(i+1)
        hy_y.DrawCopy()
        #r.gPad.SetLogy()

    c.Draw()
    c.Print("./images/EnergyDependance_Calo"+str(calo).zfill(2)+".png")
```
