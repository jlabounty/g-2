### Now lets test the histogram fitting

```python
h = r.TH1D("h","Example Fit", 1000,-100,-10)
func = r.TF1("func","gaus",0,100)
func.SetParameters(23,30,5,4)
h.FillRandom("func",10000000)

c = r.TCanvas()
h.Draw()
c.Draw()
```

```python
ding = fitHist(h,func)
```

```python
fig,ax = ding.draw()
ax[0].set_ylim(0.1,1000)
ax[0].set_yscale("log")
plt.show()
```

```python
xs = np.random.normal(size=100000)
print(xs)

ding = plt.hist(xs,bins=320, range=(-5,5))
```

```python
func = r.TF1("func","gaus",-100,100)
dong = fitHist(ding,func,binErrors=True)
```

```python
fig,ax = dong.draw()
ax[0].set_ylim(0.1,1500)
plt.show()
```
