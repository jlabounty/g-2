```python
import ROOT as r
import random
import numpy as np
```

```python
func = r.TF1("func","[0] + [1]*x + [2]*(x - [3])*(x - [3])+ [4]*(x - [5])*(x - [5])*(x - [5])",0,11)
func.SetParameters(0,1,-6,5,0.8,4)
c =r.TCanvas()
func.Draw()
c.Draw()
```

```python
nums = [x for x in np.linspace(0,11,1000)]
```

```python
vals = []
for x in nums:
    vi = func.Eval(x) + random.gauss(0, 2)
    vals.append(vi)
```

```python
gr = r.TGraph()
for i in range(len(nums)):
    gr.SetPoint(i, nums[i], vals[i])
```

```python
c =r.TCanvas()
func.Draw()
gr.Draw("p SAME")
c.Draw()
```

```python
c = r.TCanvas()

ymin = -200
ymax = 75

x1 = 0
x1f = 6

x2 = 5
x2f = 11

gr.SetTitle("Overlapping Subsets Example")
gr.GetXaxis().SetRangeUser(0,11)
gr.GetYaxis().SetRangeUser(ymin,ymax)
gr.Draw("ap")

box1 = r.TBox(x1,ymin,x1f,ymax)
box1.SetFillColorAlpha(38,0.2)
box1.Draw("same")
fit1 = r.TF1("fit1","pol1",x1,x1f)
fit1.SetLineColor(4)
gr.Fit(fit1,"R")
fit1.Draw("SAME")

box2 = r.TBox(x2,ymin,x2f,ymax)
box2.SetFillColorAlpha(45,0.2)
box2.Draw("SAME")
fit2 = r.TF1("fit1","pol1",x2,x2f)
fit2.SetLineColor(2)
gr.Fit(fit2,"R")
fit2.Draw("SAME")

leg = r.TLegend(0.7,0.1,0.9,0.2)
leg.AddEntry(fit1, "Fit to Blue Region","l")
leg.AddEntry(fit2, "Fit to Red Region","l")
leg.Draw(
)
#c.SetGrid()
c.Draw()
```
