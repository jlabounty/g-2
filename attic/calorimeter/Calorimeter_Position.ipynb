```python
x = 0
t1 = 0
timerange = 4370
t2 = t1+timerange
num = 60
list = []
listtime = []
while x < num:
    string = "caloNum == 7 && time>"+str(t1)+" && time < "+str(t2)
    list.append(string)
    listtime.append(str(t1/1000)+"#mus - "+str(t2/1000)+"#mus")
    x = x+1
    t1 = t2
    t2 = t2+timerange
print(list)
print(listtime)
```

```python
x = 0
e1 = 0
energyrange = 100
e2 = e1+energyrange
t1 = 15*1000/1.25
num = 30
list2 = []
listenergy = []
while x < num:
    string = "caloNum == 7 && energy > "+str(e1)+" && energy < "+str(e2)+" && time>"+str(t1)
    list2.append(string)
    listenergy.append(str(e1)+" MeV - "+str(e2)+" MeV")
    x = x+1
    e1 = e2
    e2 = e2+energyrange
print(list2)
print(listenergy)
```

```python
c = r.TCanvas()
c.SetLogz()
i = 0
while i < num:
    t.Draw("x:y",list2[i],"colz")
    t.SetTitle(listenergy[i])
    ding = "./positionSpectrum/"+str(i)+"byEnergy.png"
    c.Print(ding)
    i = i+1
c.Draw()
```
