```python
import ROOT as r
r.gStyle.SetOptStat(0)
```

```python
f = r.TFile("./gm2offline_ana_7893458_15937.00027.root")
f.cd("clusterTree")
f.ls()
```

```python
t = f.Get("clusterTree").Get("clusters")
```

```python
t.Print()
```

```python
x = 0
t1 = 0
timerange = 4370
t2 = t1+timerange
num = 60
list = []
listtime = []
while x < num:
    string = "caloNum == 7 && time>"+str(t1)+" && time < "+str(t2)+" && energy<4500"
    list.append(string)
    listtime.append(str(t1/1000)+"#mus - "+str(t2/1000)+"#mus")
    x = x+1
    t1 = t2
    t2 = t2+timerange
print(list)
print(listtime)
```

```python
c = r.TCanvas()
c.SetLogy()
h = r.TH1D("h",listtime[0],450,0,4500)
h.SetMarkerColor(1)
h.SetMarkerStyle(8)
h.SetLineColor(1)
h.SetMarkerSize(.25)
t.Draw("energy>>h",list[0],"goff")
h.Draw("E0 P0")
leg = r.TLegend(0.1,0.7,0.48,0.9)
leg.AddEntry("h",listtime[0],"pl")
leg.SetNColumns(2)
i = 0
h2 = r.TH1D("h2","h2",450,0,4500)
while i < 10:
    i = i+1
    h2.SetMarkerColor(i+2)
    h2.SetLineColor(i+2)
    h2.SetMarkerStyle(8)
    h2.SetMarkerSize(.25)
    h2.SetTitle(listtime[i])
    t.Draw("energy>>h2",list[i],"goff")
    h2.DrawCopy("P0 E0 SAME")
    leg.AddEntry(h2,listtime[i],"pl")
#leg.Draw("SAME")
c.BuildLegend()
c.Draw()
```

```python
c2 = r.TCanvas()
c2.SetLogy()
i = 0
h2 = r.TH1D("h2","h2",450,0,4500)
h2.SetMarkerColor(2)
h2.SetLineColor(2)
h2.SetMarkerStyle(8)
h2.SetMarkerSize(.25)
while i < len(list):
    h2.SetTitle(listtime[i])
    t.Draw("energy>>h2",list[i],"goff")
    h2.DrawCopy("P0 E0")
    leg = r.TLegend(0.7,0.8,0.9,0.9)
    leg.AddEntry(h2,listtime[i],"pl")
    leg.Draw("SAME")
    #c.BuildLegend()
    c2.Draw()
    ding = "./energySpectrum/"+str(i)+".png"
    c2.Print(ding)
    i = i+1
```

```python
c = r.TCanvas()
c.SetLogy()
h = r.TH1D("h",listtime[0],450,0,4500)
h.SetMarkerColor(1)
h.SetMarkerStyle(8)
h.SetLineColor(1)
h.SetMarkerSize(.25)
t.Draw("energy>>h",list[0],"goff")
h.Scale(1/(h.GetEntries()))
h.Draw("E0 P0")
c.Draw()
```

```python
c2 = r.TCanvas()
c2.SetLogy()
i = 0
h2 = r.TH1D("h2","h2",450,0,4500)
h2.SetMarkerColor(2)
h2.SetLineColor(2)
h2.SetMarkerStyle(8)
h2.SetMarkerSize(.25)
while i < len(list):
    h2.SetTitle(listtime[i])
    t.Draw("energy>>h2",list[i],"goff")
    norm = h2.GetEntries()
    h2.Scale(1/norm)
    h2.DrawCopy("P0 E0")
    leg = r.TLegend(0.7,0.8,0.9,0.9)
    leg.AddEntry(h2,listtime[i],"pl")
    leg.Draw("SAME")
    #c.BuildLegend()
    c2.Draw()
    ding = "./energySpectrum/"+str(i)+"_norm.png"
    c2.Print(ding)
    i = i+1
```
