### Lonely clusters energy vs. time

```python
f1 = r.TFile("./data/LonelyClusters_energyWest_vs_timeWest_9day_Calo5.root")
f1.ls()
```

```python
c1 = f1.Get("c1")
c1.Draw()
```

```python
print([x for x in c1.GetListOfPrimitives()])
h1 = c1.GetPrimitive("h").Clone("h1")
```

```python
c = r.TCanvas()
h1.Draw("colz")
c.Draw()
```

```python
c = r.TCanvas()
h1.ProjectionY().Draw("colz")
c.SetLogy()
c.Draw()
```

```python
h1.GetXaxis().SetRangeUser(30,40)
c.Draw()
```

```python
f2 = r.TFile("./data/LonelyClusters_energyWest_vs_timeWest_9day_NoCalo5.root")
f2.ls()
```

```python
c2 = f2.Get("c1")
c2.Draw()
```

```python
print([x for x in c2.GetListOfPrimitives()])
h2 = c2.GetPrimitive("h").Clone("h2")
h2.SetLineColor(2)
```

```python
c = r.TCanvas()

h2y = h2.ProjectionY().Clone("h2y")
h2y.Draw("")
h2y.SetTitle("Energy of Lonely Recon West Clusters (9day)")

h1.ProjectionY().Draw("same")
c.SetLogy()

leg = r.TLegend(0.6,0.8,0.9,0.9)
leg.AddEntry(h1,"Only Calo 5", "fl")
leg.AddEntry(h2,"All Other Calos", "fl")
leg.Draw()

c.Draw()
```
