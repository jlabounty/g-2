```python
import ROOT as r
import math
from BlindersPy3 import Blinders
from BlindersPy3 import FitType
r.gStyle.SetOptStat(0)
r.gStyle.SetOptFit(1111)
import matplotlib.pyplot as plt
import csv
import pandas

import seaborn as sns
sns.set(style="ticks", color_codes=True)
import numpy as np
import matplotlib.pyplot as plt
import matplotlib 
matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20) 
import os
from io import StringIO
import urllib.request

%matplotlib inline
```

```python
truevec = []
lastsubrunvec = []

for i in range(1000):
    ding = urllib.request.urlopen("http://dbdata0vm.fnal.gov:9091/QE/gm2/dev/app/SQ/query?dbname=postgres_dev&t=nearline_sync_energy_constants&w=validfromrun:16355&w=calonum:24&w=isvalid:t&o=-validfromrun&l=1")
    ding = ding.read()
    
    f = StringIO(str(ding))
    reader = csv.reader(f, delimiter=',')
    for item in reader:
        if(item[12]=="True"):
            truevec.append(1)
        else:
            truevec.append(0)
            
        lastsubrunvec.append(int(item[14][:-3]))
```

```python
plt.plot(truevec,'.:')
plt.ylabel("True (1) or False (0)")
plt.xlabel("Attempt")
plt.title("Database DDOS")
plt.show()

plt.plot(lastsubrunvec,'.:')
plt.title("Database DDOS")
plt.ylabel("Last Subrun")
plt.show()
```
