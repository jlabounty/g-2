```python
import ROOT as r
import math
from BlindersPy3 import Blinders
from BlindersPy3 import FitType
r.gStyle.SetOptStat(0)
r.gStyle.SetOptFit(1111)
import matplotlib.pyplot as plt
import matplotlib
```

```python
c = r.TCanvas()
c.Divide(1,2)
c.cd(1)
t.Draw("energy","laserHit==0 && time>30*1000/1.25 && time < 650*1000/1.25 && caloNum==1")
c.cd(2)
t.Draw("energy","laserHit==1 && time>30*1000/1.25 && time < 650*1000/1.25 ")
c.Draw()
```

```python
listVec = []

for calo in range(1,25):
    h = r.TH1D()
    t.Draw("xtalNum>>h","caloNum == "+str(calo)+" && energy>100 && energy<6000 && time>30*1000/1.25 && time < 650*1000/1.25 &&laserHit==0","goff")
    hi2 = r.gDirectory.Get("h")
    ding = []
    for i in range(1,55):
        ding.append(hi2.GetBinContent(i))
    
    ding.reverse() #now the 54th calo is first in like, and thus will be the upper left corner.
    list1 = []
    list2 = []
    for i in range(54):
        list2.append(ding[i]) 
        if (i+1) % 9 == 0: #split into 9 column blocks
            list1.append(list2)
            list2 = []
            
    listVec.append(list1)
    print(list1)
       
    plt.imshow(list1,cmap='summer',vmin=0.1)
    plt.title("Raw Counts in Calo "+str(calo))
    plt.xlabel('x position (crystals)')
    plt.ylabel('y position (crystals)')
    plt.colorbar()
    plt.show()
```

```python
for i,list1 in enumerate(listVec):
    print(list1)
    cmap = plt.cm.viridis
    cmap.set_under(color='black')    
    cmap.set_bad(color='black')    
    cs = plt.imshow(list1,vmin=4000, vmax=70000,norm=matplotlib.colors.LogNorm(),cmap=cmap,interpolation='none')
    plt.title("Raw Counts in Calo "+str(i+1)+" (100 MeV < E < 6000 MeV, \n 30$\mu$s < t < 650 $\mu$s)")
    plt.xlabel('x position (crystals)')
    plt.ylabel('y position (crystals)')
    plt.colorbar()
    plt.savefig("../images/SingleHits/calo"+str(i+1).zfill(2)+"_EnergyBound_100_6000_crystalHits_timeCut.png",bbox_inches='tight')
    plt.show()
```

```python
listVecLaserEnergyProfile = []

for calo in range(1,25):
    h = r.TH1D()
    h2 = r.TH2D()
#    t.Draw("xtalNum:energy>>h2","caloNum == "+str(calo)+" && energy>100 && energy<6000 && time>30*1000/1.25 && time < 650*1000/1.25 && laserHit==0","goff")
    t.Draw("xtalNum:energy>>h2","caloNum == "+str(calo)+" && laserHit==1","goff")
    hi2 = r.gDirectory.Get("h2")
    c = r.TCanvas()
    hi2.Draw()
    c.Draw()
    h = hi2.ProfileY()
    c2 = r.TCanvas()
    h.Draw()
    c2.Draw()
    
    ding = []
    for i in range(1,55):
        ding.append(h.GetBinContent(i))
    
    ding.reverse() #now the 54th calo is first in like, and thus will be the upper left corner.
    list1 = []
    list2 = []
    for i in range(54):
        list2.append(ding[i]) 
        if (i+1) % 9 == 0: #split into 9 column blocks
            list1.append(list2)
            list2 = []
            
    listVecLaserEnergyProfile.append(list1)
    print(list1)
       
    plt.imshow(list1,cmap='summer',vmin=0.1)
    plt.title("Energy From Raw Counts in Calo "+str(calo))
    plt.xlabel('x position (crystals)')
    plt.ylabel('y position (crystals)')
    plt.colorbar().set_label('Energy (MeV)')
    plt.show()
    
    h.Delete()
    h2.Delete()
    hi2.Delete()
```

```python
for i,list1 in enumerate(listVecEnergyProfile):
    print(list1)
    cmap = plt.cm.viridis
    cmap.set_under(color='black')    
    cmap.set_bad(color='black')    
#    cs = plt.imshow(list1,vmin=1000, vmax=5000,norm=matplotlib.colors.LogNorm(),cmap=cmap,interpolation='none')
    cs = plt.imshow(list1,vmin=1, vmax=800,cmap=cmap,interpolation='none')
    plt.title("Raw Counts in Calo "+str(i+1)+" (100 MeV < E < 6000 MeV, \n 30$\mu$s < t < 650 $\mu$s)")
    plt.xlabel('x position (crystals)')
    plt.ylabel('y position (crystals)')
    plt.colorbar().set_label('Energy (MeV)')
#    plt.savefig("../images/SingleHits/calo"+str(i+1).zfill(2)+"_EnergyBound_100_6000_crystalHits_timeCut.png",bbox_inches='tight')
    plt.show()
```
