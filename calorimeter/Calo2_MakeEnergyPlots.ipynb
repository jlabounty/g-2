```python
import ROOT as r
import math
from BlindersPy3 import Blinders
from BlindersPy3 import FitType
r.gStyle.SetOptStat(0)
r.gStyle.SetOptFit(1111)
import matplotlib.pyplot as plt
import csv
import pandas

import seaborn as sns
sns.set(style="ticks", color_codes=True)
import numpy as np
import matplotlib.pyplot as plt
import matplotlib 
matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20) 

from scipy import stats

%matplotlib inline
```

```python
f = r.TFile("./data/Calo2Projections.root")
f.cd()
f.ls()
```

```python
def GetKeyNames( self, dir = "" ):
        self.cd(dir)
        return [key.GetName() for key in r.gDirectory.GetListOfKeys()]
r.TFile.GetKeyNames = GetKeyNames

keyList = f.GetKeyNames("")
print ("\nKeys in file:", keyList)
#h = r.TH3D()
#f.GetObject(keyList[1], h)
```

```python
projectionVec = []
c = r.TCanvas()
for i in range(24):
    hi = r.TH3D()
    f.GetObject(keyList[i],hi)
    projI = hi.Project3DProfile("yz")
    projI.GetZaxis().SetRangeUser(0,3000)
    projI.Draw("colz")
    projI.GetXaxis().SetTitle("x position of cluster mean")
    projI.GetYaxis().SetTitle("y position of cluster mean")
    projI.SetTitle("Calo "+str(i+1)+" Energy Projection (1700 < E < 6000)")
    c.Print("./energyInCalos/energyCut/Calo"+str(i+1).zfill(2)+"_energyCut.png")
    c.Draw()
```
