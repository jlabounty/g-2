```python
eventTreeExists = False
try:
    t = f.Get("farline").Get("eventTree")
    t.Print()
    
    eventTreeExists = True
    
except:
    print("ERROR: eventTree not found. Likely using results_histOnly.root")
```

```python
if(eventTreeExists):
    for i in range(1,2): 
        print("Starting calo", i)
        c3 = r.TCanvas()
        
        t.Draw("energy:time","time > 0 && time < 6000 && energy > 40000","")
        
        c3.Draw()
        c3.Print("./images/"+prodString+"/"+prodString+"_laserEnergyPulse_calo"+str(i).zfill(2)+".png")
        c3.Print("./images/"+prodString+"/"+prodString+"_laserEnergyPulse_calo"+str(i).zfill(2)+".root")
```

```python
if(eventTreeExists):
    for i in range(1,2): 
        print("Starting calo", i)
        c3 = r.TCanvas("c3","c3",4500,4500)
        c3.Divide(4,5)
        
        tStart = 0 #us
        tEnd = 700 #us
        tDiv = 43.7 #us
        nDivs = int(np.ceil( (tEnd - tStart)/tDiv ))
        print(nDivs,"Divisions")
        for div in range(nDivs):
            c3.cd(div+1)
            
            ti = (tDiv * div) * 1000 / 1.25
            tf = (tDiv * (div+1)) * 1000 / 1.25
            print("     ", div, ti, tf)
                        
            cutString = "caloNum=="+str(i)+" && time < "+str(tf)+" && time > "+str(ti)
            print("     ", cutString)
            
            hi = r.TH1D( "hi", cutString, 250, 0, 4000)
            
            t.Draw("energy>>hi",cutString,"goff")
            hi.DrawCopy("colz")
            c3.Draw()
            
            hi.Delete()
        
        c3.Print("./images/"+prodString+"/"+prodString+"_energySpectrum_calo"+str(i).zfill(2)+".png")
        c3.Print("./images/"+prodString+"/"+prodString+"_energyPulse_calo"+str(i).zfill(2)+".root")
```

```python
if(eventTreeExists):
    for i in range(1,2): 
        print("Starting calo", i)
        c3 = r.TCanvas("c3","c3",4500,2000)
        ncol = 4
        nrow = 2
        c3.Divide(ncol,nrow)
        minEnergy = 0 #MeV
        maxEnergy = 3000 #MeV
        ndivs = ncol*nrow #how many divisions should be break this into
        Ebin = float(maxEnergy - minEnergy) / ndivs
        for division in range(ndivs):
            
            print("      division:", division)
            Elow = Ebin*division
            Ehigh = Ebin*(division+1)
            c3.cd(division + 1)
            cutString = "caloNum=="+str(i)+" && energy < "+str(Ehigh)+" && energy > "+str(Elow)
            print("     ", cutString)
            
            hi = r.TH2D("hi", cutString, 900, 0, 9, 600, 0, 6)
            t.Draw("y:x>>hi",cutString,"goff")
           
            hi.DrawCopy("colz")
            hi.Delete()
            
            r.gPad.SetLogz()
            
        c3.Draw()
        c3.Print("./images/"+prodString+"/"+prodString+"_hitDistribution_calo"+str(i).zfill(2)+"_Energy_"+str(Elow)+"_"+str(Ehigh)+".png")
        c3.Print("./images/"+prodString+"/"+prodString+"_hitDistribution_calo"+str(i).zfill(2)+"_Energy_"+str(Elow)+"_"+str(Ehigh)+".root")
```
