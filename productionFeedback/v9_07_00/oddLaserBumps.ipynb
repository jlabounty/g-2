```python
df = pandas.read_csv("./ding.csv",header=None)
df.head()
newcol = []

for j in range(df.shape[0]):
    test = df[3][j]
    test = test[1:len(test)-1]
    test = test.split(',')
    for i,ding in enumerate(test):
        test[i] = float(ding)
    newcol.append(test)
    
df[8] = pandas.Series(newcol,index=df.index)
df.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 5))

for i, row in df.iterrows():
    if(row[5]=='t'):
        row1 = row[:]
        print(row1)
        break


for i, row in df.iterrows():
    plt.plot([row[8][j]/row1[8][j]*100 for j in range(len(row[8]))])
#    if row[8][0] < 1.2:
#        print(row)
plt.ylabel("driftCorrection Constant (% of Reference Entry)")
plt.xlabel("Crystal Number")
plt.title("Long Term Gain Calibration Constants vs. Xtal Number for Calo 24 (runs 16351 - 16358)")
plt.show()
```
