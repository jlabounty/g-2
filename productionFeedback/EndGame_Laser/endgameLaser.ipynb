```python
gainFunc = r.TF1("gainFunc","[0] + [1]*(1 - [2]*TMath::Exp(-1*x / [3]))",0,700,4)
gainFunc.SetParameters(87400, 100, 1, 100)
c = r.TCanvas()
gainFunc.Draw()
c.Draw()
```

```python
c = r.TCanvas()
projections[0].Fit(gainFunc,"REMB")
projections[0].Draw()
c.Draw()
```

```python
h = r.TH1D("h","Residuals From Gain Fit; Time [#mus]; Energy [MeV]",700,0,700)
for i in range(h.GetNbinsX()+1):
    content = projections[0].GetBinContent(i)
    funcEval = gainFunc.Eval( h.GetBinCenter(i) )
    #print(content, funcEval)
    if(content > 0.1):
        h.SetBinContent(i, content - funcEval)
c = r.TCanvas()
h.Draw()
c.Draw()
```

```python
fft = fourierXformWiggle( h, gainFunc, 0, 600, "Fourier Transform of Residuals").Clone("fft")
```

```python
c = r.TCanvas()
fft.Draw("hist")
c.SetLogy()
c.Draw()
```

```python
caloHists_EnergyTimeRun = []
for calo in range(1,2):
    print("Starting calo", calo)
    h = r.TH3D("h","Energy of Lasers in Calo "+str(calo)+" vs. Time in Fill; Time [#mus]; Energy [MeV] ; Run Number", 
              700,0,700,
              800,40000,120000,
              50, 16930, 16980)
    ch.Draw("runNum:energy:time*1.25/1000.>>h","energy > 10000 && caloNum == "+str(calo),"goff")
    caloHists_EnergyTimeRun.append( h.Clone("h_"+str(calo)))
```

```python
c = r.TCanvas("c","c",2000,1400)
ding = caloHists_EnergyTimeRun[0].Project3DProfile("yx").Clone()
ding.GetZaxis().SetRangeUser(16930,16980)
ding.Draw("colz")
c.Draw()
```

```python
profiles_byRun = []
runDiv = 5
for run in range(16930, 16980, runDiv):
    run2 = run+runDiv
    caloHists_EnergyTimeRun[0].GetZaxis().SetRangeUser(run, run2)
    ding = caloHists_EnergyTimeRun[0].Project3D("yx").Clone("h_"+str(run))
    ding.SetTitle("Runs: "+str(run)+" - "+str(run2))
    profiles_byRun.append(ding)
```

```python
print(profiles_byRun)
```

```python
c = r.TCanvas("c","c",4000,2000)
c.Divide(5,2)
profiles = []
for i in range(len(profiles_byRun)):
    c.cd(i+1)
    ding = profiles_byRun[i].ProfileX().Clone(str(i))
    ding.GetYaxis().SetRangeUser(80000,90000)
    profiles.append(ding)
    profiles[i].Draw("colz")
c.Draw()
```
