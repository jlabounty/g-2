```python
unevenCounter = 0
for i, e in enumerate(t):
    try:
        if( ( e.crystalsFull.size() != e.crystalsTrunc.size() ) or (e.crystalsTrunc.size() > 1) ):
            continue
            
        fullXtals = e.crystalsFull[0]
        truncXtals = e.crystalsTrunc[0]
        
        if( fullXtals.size() == truncXtals.size() ):
            continue
            
        if(unevenCounter > 20):
            break
            
        unevenCounter += 1
        
        print("----------------------------------------")
        
        print("     ", [[y for y in x] for x in e.crystalsFull])
        print("     ", [[y for y in x] for x in e.crystalsTrunc])
        
        xtalsFull = [x for x in e.crystalsFull[0]]
        xtalsTrunc = [x for x in e.crystalsTrunc[0]]
        
        #print(xtalsFull)
        #print(xtalsTrunc)
        
        imshowCompare( toMap(xtalsFull), toMap(xtalsTrunc), True, False )
        
        fullEnergiesUncalib = e.energiesFullUncalib
        print("     ", [x for x in fullEnergiesUncalib])
        truncEnergiesUncalib = e.energiesTruncUncalib
        print("     ", [x for x in truncEnergiesUncalib])
            
        fullEnergies = e.energiesFull
        print("     ", [x for x in fullEnergies])
        truncEnergies = e.energiesTrunc
        print("     ", [x for x in truncEnergies])
    except:
        continue
```
