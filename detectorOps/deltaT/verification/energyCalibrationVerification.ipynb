```python
c = r.TCanvas("c","c",2400,1500)
c.Divide(6,4)
for calo in range(1,25):
    c.cd(calo)
    h = r.TH1D("h","Energy in Calo "+str(calo)+"; Energy [MeV]; N",300,0,6000)
    t.Draw("energy>>h",
      "time > 24000 && caloNum == "+str(calo),
      "goff")
    h.Scale(1/h.Integral())
    h.GetXaxis().SetRangeUser(0,3800)
    h.GetYaxis().SetRangeUser(0.00005,0.05)
    h.SetFillColorAlpha(38,0.2)
    h.DrawCopy("hist")
    r.gPad.SetLogy()
    
    
    h2 = r.TH1D("h2","Energy in Calo "+str(calo),300,0,6000)
    t2.Draw("energy>>h2",
      "time > 24000 && caloNum == "+str(calo),
      "goff")
    h2.Scale(1/h2.Integral())
    h2.GetXaxis().SetRangeUser(0,3800)
    h2.GetYaxis().SetRangeUser(0.00005,0.5)
    h2.SetFillColorAlpha(2,0.2)
    h2.DrawCopy("hist same")
    r.gPad.SetLogy()
    
leg = r.TLegend(0.1,0.1,0.5,0.3)
leg.AddEntry(h,"Original","lf")
leg.AddEntry(h2,"Post Change","lf")
leg.Draw("SAME")
     
c.Draw()
```
