```python
c = r.TCanvas()
#calo = 9
allCaloDeltaTs = []
#for calo in range(1,25):
for calo in range(1,25):
    print("Starting calo", calo)
    h = r.TH3I("h","Map of #bar{#deltat} by xtal; x; y; delta t", 9,0,9,6,0,6,100,0,10)
    t.Draw("deltaT:y:x>>h","time > 24000 && nHit == 1 && coinNum == 1 && caloNum == "+str(calo),"goff")
    allCaloDeltaTs.append(h.Clone("h_"+str(calo)))
h.Draw("lego2z")
c.Draw()
```

```python
c = r.TCanvas()
h_xy = h.Project3DProfile("yx").Clone("h_xy")
h_xy.SetTitle("Map of #bar{#deltat} by xtal in calo "+str(calo)+"; x [xtal]; y [xtal]")
h_xy.Draw("colz text")
c.Draw()
```

```python
c = r.TCanvas("c","c", 4800, 2000)
c.Divide(6,4)
for calo in range(1,25):
    c.cd(calo)
    r.gPad.SetRightMargin(0.13)
    h_xy = allCaloDeltaTs[calo-1].Project3DProfile("yx").Clone("h_xy")
    mean = h_xy.GetMean(3)
    #print(mean)
    h_xy.GetZaxis().SetRangeUser(mean-1, mean+1)
    h_xy.SetTitle("Map of #bar{#deltat} by xtal in calo "+str(calo)+"; x [xtal]; y [xtal]; #bar{#deltat} [c.t.]")
    h_xy.DrawCopy("colz text")
c.Draw()
```

```python
c = r.TCanvas("c","c", 4800, 2000)
c.Divide(6,4)
for calo in range(1,25):
    c.cd(calo)
    r.gPad.SetRightMargin(0.13)
    h_xy = allCaloDeltaTs[calo-1].Project3D("zx").Clone("h_xy")
    mean = h_xy.GetMean(3)
    #print(mean)
    #h_xy.GetZaxis().SetRangeUser(mean-1, mean+1)
    h_xy.SetTitle("Map of #bar{#deltat} by xtal in calo "+str(calo)+"; x [xtal]; y [xtal]; #bar{#deltat} [c.t.]")
    h_xy.DrawCopy("colz")
    r.gPad.SetGrid()
    r.gPad.SetLogz()
c.Draw()
```
