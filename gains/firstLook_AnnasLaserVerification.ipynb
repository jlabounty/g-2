```python
f1 = r.TFile("./data/gm2gain_ana.root")
f1.cd("LaserPlotsSTDPs/Calo01")
f1.ls()
```

```python
print([ (x.ReadObj()) for x in f1.Get("LaserPlotsSTDPs").Get("allcalos").GetListOfKeys()])
```

```python
profiles = []
for x in f1.Get("LaserPlotsSTDPs").Get("Calo02").GetListOfKeys():
    if("TProfile" in str(type(x.ReadObj()))):
        profiles.append( x.ReadObj() )
print(profiles)
```

```python
nrows = 2
c = r.TCanvas("c","c",int(600*len(profiles)/nrows),600*nrows)
c.Divide(int(np.ceil(len(profiles)/nrows)),nrows)
for i, pi in enumerate(profiles):
    c.cd(i+1)
    pi.Draw()
    print(pi.GetEntries())
c.Draw()
```
