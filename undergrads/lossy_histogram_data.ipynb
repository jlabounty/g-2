```python
from si import *
```

```python
bins = [10,20,50,100,1000,10000,10000000]
```

```python
example_sine = r.TF1("example_sine","[0]*TMath::Sin([1]*x + [2]) + [3]*TMath::Sin([4]*x + [5]) + [6]*TMath::Sin([7]*x + [8]) +2", 0,50,9)
example_sine.SetParameters(1,1,0,
                           0.5,.51,2,
                           0.21,10,1)
example_sine.SetNpx(10000)

c = r.TCanvas()
example_sine.Draw()
example_sine.GetXaxis().SetRangeUser(0,40)
c.Draw()
```

```python
hists = []
for bini in bins:
    hi = r.TH1I("hi","Histogram with "+str(bini)+" bins", bini, 0, 50)
    for i in range(1000000):
        hi.Fill(example_sine.GetRandom())
    hists.append(hi.Clone(str(bini)))
    
```

```python
c = r.TCanvas("c","c",1200,1000)
c.Divide(2,int(len(hists)/2))
for i, hi in enumerate(hists):
    c.cd(i+1)
    hi.Draw()
c.Draw()
```
