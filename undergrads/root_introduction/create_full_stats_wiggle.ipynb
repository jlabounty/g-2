```python
f = r.TFile("../../slidingScaleFits/data/EndGame4042BSeeds0.root")
f.cd("clustersAndCoincidencesRand")
f.ls()
```

```python
h = f.Get("clustersAndCoincidencesRand/clusters").Clone()
h.SetDirectory(0)
```

```python
c = r.TCanvas("x","x",1600,500)
c.Divide(3)
for i, xi in enumerate(['x', 'y', 'z']):
    c.cd(i+1)
    h.Project3D(xi).Draw()
    r.gPad.SetLogy()
    r.gPad.SetGrid()
c.Draw()
```

```python
h2 = h.Project3D("yx").Clone("energy_vs_time")
c = r.TCanvas()
h2.Draw("colz")
c.SetLogz()
c.Draw()
```

```python
fout = r.TFile("./output_wiggle_TH2.root", "RECREATE")
h2.Write("energy_vs_time")
fout.Write()
fout.Close()
```
