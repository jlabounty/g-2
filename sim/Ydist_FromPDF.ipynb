### Now lets put this into a loop

```python
histVec_ry = []
histVec_r = []
histVec_y = []

for calo in range(1,25):
    print("Starting Calo:", calo)
    fi = r.TFile("./data/PDF_MuonUnscaled_Energies_1700_3500_Calo"+str(calo)+"_May16.root")
    #fi.ls()
    ci = fi.Get("c")
    #print([c for c in ci.GetListOfPrimitives()])
    hi = ci.GetPrimitive("PDF_AllEnergies").Clone("hi")
    hi_ry = hi.Project3D("yx").Clone()
    hi_ry_scaled = hi_ry.Clone("hi_ry_scaled")
    hi_ry_scaled.Divide(hmu_ry)
    histVec_ry.append( hi_ry_scaled.Clone("h"+str(calo)+"_ry_scaled") )
    histVec_ry[calo - 1].SetDirectory(0)
    
    hi_y = hi.Project3D("y").Clone()
    hi_y_scaled = hi_y.Clone("hi_y_scaled")
    hi_y_scaled.Divide(hmu_y)
    histVec_y.append( hi_y_scaled.Clone("h"+str(calo)+"_y_scaled") )
    histVec_y[calo - 1].SetDirectory(0)

    
    hi_r = hi.Project3D("x").Clone()
    hi_r_scaled = hi_r.Clone("hi_r_scaled")
    hi_r_scaled.Divide(hmu_r)
    histVec_r.append( hi_r_scaled.Clone("h"+str(calo)+"_r_scaled") )
    histVec_r[calo - 1].SetDirectory(0)
```

```python
roughThetas = [0.11272777572964064, 0.38704268778924433, 0.630334955791008, 0.9882987408784784, 1.1718751132875604, 1.435603925064294, 
               1.6943768479695187, 1.9531923566210057, 2.2134087184367677, 2.475283362118464, 2.7397174588150497, 2.9979804262428855, 
               3.272053406490232, 3.606752905906422, 3.790081889744413, 4.047968629057927, 4.3078791723918215, 4.654521173053848, 
               4.8274337538242085, 5.102626352202925, 5.339031467575712, 5.607840660438622, 5.884482084626629, 6.138585371556525]

roughThetaDeg = [x*180/r.TMath.Pi() for x in roughThetas]
print(roughThetaDeg)
```

```python
for calo in range(1,25):
    c = r.TCanvas("c","c",4000,2000)
    c.Divide(2,2)

    ymin = 0.002
    ymax = 0.007
    title = "Acceptance in Calo "+str(calo)+" [1700 < E < 3500 MeV]"

    h10_ry_scaled = histVec_ry[calo-1]
    h10_r_scaled = histVec_r[calo-1]
    h10_y_scaled = histVec_y[calo-1]
    
    c.cd(1)
    h10_ry_scaled.SetTitle(title)
    h10_ry_scaled.Draw("colz")
    h10_ry_scaled.SetFillColorAlpha(38,0.5)
    h10_ry_scaled.GetZaxis().SetRangeUser(ymin,ymax)
    r.gPad.SetRightMargin(1)
    h10_ry_scaled.GetZaxis().SetTitleOffset(0)

    c.cd(2)
    h10_y_scaled.SetTitle(title+" --- y")
    h10_y_scaled.SetFillColorAlpha(38,0.5)
    h10_y_scaled.GetYaxis().SetRangeUser(ymin,ymax)
    h10_y_scaled.Draw("hbar")

    c.cd(3)
    h10_r_scaled.SetFillColorAlpha(38,0.5)
    h10_r_scaled.SetTitle(title+" --- r")
    h10_r_scaled.GetYaxis().SetRangeUser(ymin,ymax)
    h10_r_scaled.Draw("")
    
    c.cd(4)
    h = r.TH2D("h","Calo "+str(calo),10,-2,2,10,-2,-2)
    h.Draw("a")
    
    caloLoc = 90 - roughThetaDeg[calo - 1]
    el0 = r.TEllipse(0.5,0.45,0.25,0.45)
    el0.SetLineColor(1)
    el0.SetLineWidth(6)
    el0.Draw("SAME")
    
    el1 = r.TEllipse(0.5,0.45,0.25,0.45,caloLoc,90)
    el1.SetLineColor(2)
    el1.SetLineWidth(6)
    el1.Draw("same")

    c.Draw()
    c.Print("./images/PDF_ry_y_r_combined_Calo"+str(calo).zfill(2)+".png")
    c.Print("./images/PDF_ry_y_r_combined_Calo"+str(calo).zfill(2)+".root")
```

### Fit this with a convolution of a box and a gaussian because I want to watch the world burn

```python
f3 = r.TF1("f3",
#           "[0]*TMath::Sqrt(TMath::Pi() / 2.0)*[1]*( TMath::Erf(([2] + x - 2*[3])/(TMath::Sqrt(2)*[1])) + TMath::Erf(([2] - x + 2*[3])/(TMath::Sqrt(2)*[1])) )", 
           "[0]/2*( TMath::Erf(([2] + x - 2*[3])/(TMath::Sqrt(2)*[1])) + TMath::Erf(([2] - x + 2*[3])/(TMath::Sqrt(2)*[1])) )", 
           -45, 45, 4)
f3.SetParameters(1,1,1,1)
f3.SetParNames("A","#sigma_{gaus}","r_{square}","#bar{x}")
f3.SetNpx(2000)
f3.SetLineColor(6)
```

### And now the radial directions

```python
rMeanVec = []
rStdVec = []
parsVec_rlin = []
parsVec_rquad = []

chisqvec_rlin = []
chisqvec_rquad = []

linVec = []
quadVec = []

f2 = r.TF1("f2","pol1",7112-45,7112+45)
f2.SetParameters(0.001,-1)

f3 = r.TF1("f3","pol2",7112-45,7112+45)
#f3 = r.TF1("f3","[0] + [1]*x + [2]*(x - [3])*(x - [3])",7112-45,7112+45)
f3.SetParLimits(2,-1,0)
f3.SetParameter(2,-0.000001)

for i, hi in enumerate(histVec_r):
    #print(hi)
    rMeanVec.append( hi.GetMean() )
    rStdVec.append( hi.GetStdDev() )
    
    f3.SetParameters(-16, 0.001, -0.00000001)
    #f3.SetParameters(-20, -0.001, -0.00000001, 0)
    
    for j in range(2):
        hi.Fit("f2","REMBQF")
        hi.Fit("f3","REMBQF")
        
    parsVec_rlin.append( [f2.GetParameter(x) for x in range(2)] )
    parsVec_rquad.append( [f3.GetParameter(x) for x in range(3)] )
    
    chisqvec_rlin.append( [f2.GetChisquare(), f2.GetNDF()] )
    chisqvec_rquad.append( [f3.GetChisquare(), f3.GetNDF()] )
    
    linVec.append(f2.Clone("f2_"+str(i)))
    quadVec.append(f3.Clone("f3_"+str(i)))
```

```python
for i, fi in enumerate(parsVec_rlin):
    print("Calo ", i+1,":", fi)
```

```python
c = r.TCanvas("c","c",3000,2000)
c.Divide(6,4)
for i, hi in enumerate(histVec_r):
    calo = i+1
    c.cd(i+1)
    hi.GetYaxis().SetRangeUser(0.003,0.007)
    hi.SetFillColorAlpha(38,0.4)
    hi.SetLineColor(1)
    if( (calo > 0 and calo < 5 ) or (calo > 6 and calo < 11) or (calo > 12 and calo < 17) or (calo > 18 and calo < 23)):
        hi.SetFillColorAlpha(30,0.7)
        hi.SetLineColor(8)
    if( (calo > 3 and calo < 8) ):
        hi.SetFillColorAlpha(6,0.3)
    if( calo == 13 or calo == 19 ):
        hi.SetFillColorAlpha(2,0.2)
    if( calo == 23 or calo == 24 ):
        hi.SetLineColor(2)
        
    if(i < 1):
        hi.Draw()
    else:
        hi.Draw()
        
    linVec[i].SetLineColor(4)
    linVec[i].Draw("SAME")
    
    pi = r.TPaveText(0.1,0.1,0.5,0.3)
    pi.AddText("hi")
    pi.Draw("")
    
c.Draw()
```

```python
chiSqOverNDF_rlin = [(x/y) for (x,y) in chisqvec_rlin]
chiSqOverNDF_rquad = [(x/y) for (x,y) in chisqvec_rquad]

chiSqDiff = [x-y for (x,y) in zip(chiSqOverNDF_rlin, chiSqOverNDF_rquad)]
```

```python
fig,ax = plt.subplots(figsize=(15,5))
plt.plot(chiSqOverNDF_rlin,"-o",label="Linear ChiSq")
plt.plot(chiSqOverNDF_rquad,"-o",label="Quadratic ChiSq")
plt.legend()
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(15,5))
plt.plot([x for x in range(1,25)], chiSqDiff,".",label="Linear - Quadratic ChiSq")
plt.legend()
plt.grid()
plt.show()
```

```python
moreLinear = []
moreQuadratic = []
for i, x in enumerate(chiSqDiff):
    if x > 0:
        moreQuadratic.append( i + 1 )
    else:
        moreLinear.append( i + 1 )
        
print("Calos which are more linear:", moreLinear)
print("Calos which are more quadratic:", moreQuadratic)
```

```python
visbar = []
for i in range(1,25):
    if i in moreLinear:
        visbar.append(-1)
    else:
        visbar.append(1)
        
fig,ax = plt.subplots(figsize=(15,5))
plt.bar([x for x in range(1,25)], visbar, label="calos")

plotRingComponents(1.1)

plt.title("Linear > 0; Quadratic < 0")
plt.xlabel("Calo Number")
plt.legend(ncol=2)
plt.grid(axis='x')
plt.show()
```
