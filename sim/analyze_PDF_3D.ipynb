### Now that we have the plots imported, lets see what we can do with them

```python
print(histVec)
rbins = histVec[0].GetNbinsX()
vbins = histVec[0].GetNbinsY()
thetabins = histVec[0].GetNbinsZ()
```

```python
for i in range(len(histVec)):
    scaledMuonDist = histVec[i].Clone("scaledMuonDist")
    scaledMuonDist.SetDirectory(0)

    for rbin in range(rbins):
        for vbin in range(vbins):
            for thetabin in range(thetabins):
                muons = hMu.GetBinContent(rbin+1,vbin+1,thetabin+1)
                positrons = scaledMuonDist.GetBinContent(rbin+1,vbin+1,thetabin+1)
                if(muons > 0):
                    scaled = positrons / muons
                else:
                    scaled = 0
                scaledMuonDist.SetBinContent(rbin+1,vbin+1,thetabin+1, scaled)
                
    c = r.TCanvas()
    scaledMuonDist.Draw("lego2z")
    c.Draw()
    c.Print("./images/ScaledPDF_RvsVvsTheta_Calo_"+str(i).zfill(2)+".png")
```
