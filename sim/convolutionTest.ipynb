```python
from standardInclude import *
```

```python
f1 = r.TF1("f1","gaus",-45,45)
f1.SetParameters(1,1,1)
f1.SetNpx(2000)
```

```python
f2 = r.TF1("f2","0 + (x > -1*[0] + [1])*(x < [0] + [1])*1",-45,45)
f2.SetParameter(0,5)
f2.SetParameter(1,1)
f2.SetNpx(2000)
f2.SetLineColor(4)
```

```python
c = r.TCanvas()
f1.Draw()
f2.Draw("SAME")
c.Draw()
```

```python
fConv = r.TF1Convolution("f1","f2",-45,45)
print(fConv.GetNpar(),"parameters in convolution")
f3 = r.TF1("f3",fConv, -45, 45, fConv.GetNpar())
f3.SetParameters(1,1,1,1,1)
f3.SetNpx(2000)
```

```python
c = r.TCanvas()
f3.SetParameters(1,1,1,5,1)
f3.Draw()
f1.Draw("SAME")
f2.Draw("SAME")
f3.GetXaxis().SetRangeUser(-10,10)
c.Draw()
```
