```python
fout = r.TFile("temp_tree_storage.root","recreate")
tcomp_partial = tcomp.CopyTree( "caloNum == 11 && deltaT < -0.5 && deltaT  > -0.54" )
print(tcomp_partial.GetEntries())
```

```python
counter = 0
for i, entry in enumerate(tcomp_partial):
    counter += 1
    if counter > 20:
        break
    print(constructUniqueID(entry), "(",counter,")")
    islandPlot(constructUniqueID(entry), teast, twest, twave, False)
    print("Energy West:", entry.energyWest,"; East:", entry.energyEast)
    print("deltaT (corrected):", entry.deltaT - entry.timeCorr)
    print("    (uncorrected:", entry.deltaT , ")")
    print("deltaX / deltaY:", entry.deltaX, "/", entry.deltaY)
    print("Time (West):", entry.timeWest, "c.t.")
    print("Time (East):", entry.timeEast, "c.t.")
    print("Crystal times (west):")
    for i in range(len(entry.crystalsWest)):
        print("     ", (entry.crystalTimesWest)[i], "c.t. in xtal", (entry.crystalsWest)[i])
    print("******************************************************************************************************")
```
