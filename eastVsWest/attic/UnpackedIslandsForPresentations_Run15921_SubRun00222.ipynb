```python
f.cd("farline")
#f.ls()
```

```python
tnon = f.Get("farline").Get("evwNonBackwardsTree")
tnon.Print()
tnon2 = f.Get("farline").Get("wveNonBackwardsTree")
#tnon2.Print()

teast = f.Get("farline").Get("eastTree")
twest = f.Get("farline").Get("eventTree")

fwave = r.TFile("../data/waveforms_15921_00222.root")
twave = fwave.Get("islands").Get("islands").Get("islandTree")
```

```python
counter = 0
for i, entry in enumerate(tnon):
    
    if(entry.energyEast > 20 and entry.fill == 33 and entry.caloNum == 14):
        counter+= 1
        print(i+1, '/', tnon.GetEntries(), "(", counter,")")
    else:
        continue
        
    if(counter > 20):
        break
    
    print("Time: ", entry.timeEast*1.25/1000, "us")
    #print(constructUniqueIDnon(entry))
    
    uniqueID = constructUniqueIDnon(entry)
    
    plotNonUnique(uniqueID, teast, twest, twave, True)
        
    islandPlot(uniqueID, teast, twest, twave, saveImage = True)
        
    print("*********************************************************************************")
```
