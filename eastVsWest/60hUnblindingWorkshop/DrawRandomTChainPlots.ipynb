### Delta x vs. Time

```python
f = r.TFile("./data/DeltaX_vs_Time_60hUnblinding.root")
f.ls()
```

```python
c1 =f.Get("c1")
print([x for x in c1.GetListOfPrimitives()])
```

```python
h = c1.GetPrimitive("h").Clone()
h.Rebin2D(1,10)
```

```python
c = r.TCanvas()
h.Draw("colz")
c.SetLogz()
#c.SetGrid()
c.Draw()
```

```python
c = r.TCanvas()
h.GetYaxis().SetRangeUser(-6,6)
h.Draw("colz")
c.SetLogz()
#c.SetGrid()
c.Draw()
```

```python
tcuts = [30,100,200,300,400]
```

```python
c = r.TCanvas()
h.Draw("COLZ")

lines = []
boundary = 3
h.GetYaxis().SetRangeUser(-1*boundary, boundary)
for i, elow in enumerate(tcuts):
    line = r.TLine(elow, -1*boundary, elow,  boundary)
    if i < 4:
        line.SetLineColor(i+1)
    else:
        line.SetLineColor(i+2)
    line.SetLineStyle(4)
    
    lines.append(line)

for line in lines:
    line.Draw("SAME")
    
c.SetLogz()
c.Draw()
```

```python
cutvec = []
nvec = []
for cut in tcuts:
    print(cut, h.GetXaxis().FindBin(cut))
    h7 = h.ProjectionY("",h.GetXaxis().FindBin(cut),-1)
    print(h7.GetEntries())
    nvec.append(h7.GetEntries())
    h7.Scale(1/(h7.GetEntries()))
    cutvec.append( h7.Clone("h"+str(cut)))
    

c = r.TCanvas()
leg = r.TLegend(0.1,0.5,0.4,0.9)

for i, h7 in enumerate(cutvec):
    print(h7)
    h7.GetXaxis().SetRangeUser(-3,3)
    #h7.GetYaxis().SetRangeUser(1,10000000)
    if i < 4:
        h7.SetLineColor(i+1)
    else:
        h7.SetLineColor(i+2)
    if i < 0.5:
        h7.Draw("HIST")
    else:
        h7.Draw("HIST SAME")
        
    leg.AddEntry(h7 ,"#deltax [t_{west} > "+str(tcuts[i])+"#mus]", "l")

        

leg.Draw()

c.SetLogy()
c.Draw()
```

```python
plt.plot(nvec,":o")
plt.yscale('log')
plt.show()
```

### Delta x vs. Time vs. Energy

```python
f = r.TFile("./data/DeltaX_vs_Time_vs_Ewest_60hUnblinding.root")
f.ls()
```

```python
c1 =f.Get("c1")
print([x for x in c1.GetListOfPrimitives()])
```

```python
h = c1.GetPrimitive("h").Clone()
```

```python
c = r.TCanvas()
h.Draw("colz")
#c.SetLogz()
#c.SetGrid()
c.Draw()
```

```python
c = r.TCanvas()
h.Project3D("yz").Draw("colz")
c.SetLogz()
#c.SetGrid()
c.Draw()
```

```python
ecuts = [200,500,1000,1700]
ecutsmax = [500,1000,1700,3100]
```

```python
c = r.TCanvas("c","c",3000,600)
c.Divide(len(ecuts))

cutvectotal = []
nvectotal = []


for canvas, ecut in enumerate(ecuts):
    ecutmax = ecutsmax[canvas]
    c.cd(canvas+1)

    h.GetZaxis().SetRangeUser(ecut,ecutmax)
    hcut = h.Project3D("yx e")
    hcut.Rebin2D(1,5)
    
    cutvec = []
    nvec = []
    
    for cut in tcuts:
        print(cut, hcut.GetXaxis().FindBin(cut))
        h7 = hcut.ProjectionY("",h.GetXaxis().FindBin(cut),-1)
        h7.SetTitle("#deltax for various times in fill ["+str(ecutmax)+" > E > "+str(ecut)+" MeV ]")
        print(h7.GetEntries())
        nvec.append(h7.GetEntries())
        h7.Scale(1/(h7.GetEntries()))
        cutvec.append( h7.Clone("h"+str(cut)))
        
    cutvectotal.append(cutvec)
    nvectotal.append(nvec)


    leg = r.TLegend(0.1,0.6,0.35,0.9)

    for i, h7 in enumerate(cutvec):
        print(h7)
        h7.GetXaxis().SetRangeUser(-5,5)
        #h7.GetYaxis().SetRangeUser(1,10000000)
        if i < 4:
            h7.SetLineColor(i+1)
        else:
            h7.SetLineColor(i+2)
        if i < 0.5:
            h7.Draw("HIST")
        else:
            h7.Draw("HIST SAME")

        leg.AddEntry(h7 ,"#deltax [t_{west} > "+str(tcuts[i])+"#mus]", "l")

    leg.Draw()

    r.gPad.SetLogy()
c.Draw()
```
