```python
df_final = pandas.read_csv("./combined_csv/run3Trends_6.csv")

df_final.head()
```

```python
df_simon = pandas.read_csv("./data/simon.csv", delimiter=",", index_col=False, skipinitialspace=True)
df_simon.columns = df_simon.columns.str.strip("\t")

df_simon['runtime'] = pandas.to_datetime( df_simon['date']+" "+df_simon['time'])
df_simon['yoke_temp_gradient'] = df_simon['yoke temp top [C]'] - df_simon['yoke temp bottom [C]']


for x in df_simon.columns:
    df_final[x] = np.nan

df_simon.head()
```

```python
df_final.head()
```

```python
for i, rowi in df_final.iterrows():
    t1 = rowi['startTime']
    t2 = rowi['stopTime']
    
    dfi = df_simon.loc[df_simon['runtime'] >= t1].loc[df_simon['runtime'] <= t2]
    
    if(dfi.shape[0] > 0):
        #print(dfi.head())
        for x in dfi.columns:
            #print(x)
            rowi[x] = list(dfi[x])[0]
            df_final.loc[i] = rowi
        #print(rowi)
        #break
```
