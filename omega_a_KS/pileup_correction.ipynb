```python
f1 = r.TFile("./data/results_clustersAndCoinc_randomized_July7.root")
f1.cd("clustersAndCoincidences")
f1.ls()

h1 = f1.Get("clustersAndCoincidences/clusters").Clone("h1")
c = r.TCanvas()
h1.Project3D("y").Draw()
c.Draw()
```

```python
f2 = r.TFile("./data/results_clustersAndCoinc_randomized_July7_pileup_corrected.root")
f2.cd("clustersAndCoincidences")
f2.ls()

h2 = f2.Get("clustersAndCoincidences/corrected").Clone("h2")
c = r.TCanvas()
h2.Project3D("y").Draw()
c.Draw()
```

```python
tlow = 100
thigh = 650

c = r.TCanvas()

h1.GetXaxis().SetRangeUser(tlow,thigh)
h1_y = h1.Project3D("y").Clone("h1_y")
h1_y.Draw()

h2.GetXaxis().SetRangeUser(tlow,thigh)
h2.SetLineColor(2)
h2_y = h2.Project3D("y").Clone("h2_y")
h2_y.Draw("same")

h2_y_inv = InvertTH1(h2_y).Clone("h1_y_inv")
h2_y_inv.SetLineColor(3)
h2_y_inv.Draw("SAME")

c.SetLogy()

c.Draw()
```

```python
f_jason = r.TFile("./data/run2c_pileup_corrected_non.root")
f_jason.ls()
h_jason = f_jason.Get("clustersAndCoincidences/corrected").Clone("h_jason")
h_jason_un = f_jason.Get("clustersAndCoincidences/uncorrected").Clone("h_jason_un")
```

```python
tlow = 50
thigh = 650

c = r.TCanvas()#"c","c",1200,1000)

h1.GetXaxis().SetRangeUser(tlow,thigh)
h1_y = h1.Project3D("y").Clone("h1_y")
h1_y.SetTitle("Comparison of Energy Spectra Before/After Pileup Correction ("+str(tlow)+"-"+str(thigh)+"#mus); Energy [MeV]; N")
h1_y.Draw()

h2.GetXaxis().SetRangeUser(tlow,thigh)
h2.SetLineColor(2)
h2_y = h2.Project3D("y").Clone("h2_y")
h2_y.Draw("same")

h2_y_inv = InvertTH1(h2_y).Clone("h1_y_inv")
h2_y_inv.SetLineColor(3)
# h2_y_inv.Draw("SAME")

h_jason.GetXaxis().SetRangeUser(tlow,thigh)
h_jason.SetLineColor(6)
h_jason_y = h_jason.Project3D("y").Clone("h_jason_y")
h_jason_y.Draw("same")

h_jason_y_inv = InvertTH1(h_jason_y).Clone("h_jason_y_inv")
h_jason_y_inv.SetLineColor(6)
# h_jason_y_inv.Draw("SAME")

h_jason_un.GetXaxis().SetRangeUser(tlow,thigh)
h_jason_un.SetLineColor(8)
h_jason_un_y = h_jason_un.Project3D("y").Clone("h_jason_un_y")
h_jason_un_y.Draw("same")


leg = r.TLegend(0.6,0.6,0.9,0.9)
leg.AddEntry(h1_y,"Uncorrected: Random")
leg.AddEntry(h2_y,"Corrected: Random")
# leg.AddEntry(h2_y_inv,"Corrected: Random (Inverse)")
leg.AddEntry(h_jason_un_y,"Uncorrected: Non-Random")
leg.AddEntry(h_jason_y,"Corrected: Non-Random")
# leg.AddEntry(h2_y_inv,"Corrected: Non-Random (Inverse)")
leg.Draw()

c.SetLogy()
c.SetGrid()

c.Draw()
```

```python
tlow = 50
thigh = 650

for tlow in range(10,150,1):
    print(tlow)

    c = r.TCanvas()#"c","c",1200,1000)

    h1.GetXaxis().SetRangeUser(tlow,thigh)
    h1_y = h1.Project3D("y").Clone("h1_y")
    h1_y.SetTitle("Comparison of Energy Spectra Before/After Pileup Correction ("+str(tlow)+"-"+str(thigh)+"#mus); Energy [MeV]; N")
    h1_y.Draw()

    h2.GetXaxis().SetRangeUser(tlow,thigh)
    h2.SetLineColor(2)
    h2_y = h2.Project3D("y").Clone("h2_y")
    h2_y.Draw("same")

    h2_y_inv = InvertTH1(h2_y).Clone("h1_y_inv")
    h2_y_inv.SetLineColor(3)
    # h2_y_inv.Draw("SAME")

    h_jason.GetXaxis().SetRangeUser(tlow,thigh)
    h_jason.SetLineColor(6)
    h_jason_y = h_jason.Project3D("y").Clone("h_jason_y")
    h_jason_y.Draw("same")

    h_jason_y_inv = InvertTH1(h_jason_y).Clone("h_jason_y_inv")
    h_jason_y_inv.SetLineColor(6)
    # h_jason_y_inv.Draw("SAME")

    h_jason_un.GetXaxis().SetRangeUser(tlow,thigh)
    h_jason_un.SetLineColor(8)
    h_jason_un_y = h_jason_un.Project3D("y").Clone("h_jason_un_y")
    h_jason_un_y.Draw("same")


    leg = r.TLegend(0.6,0.6,0.9,0.9)
    leg.AddEntry(h1_y,"Uncorrected: Random")
    leg.AddEntry(h2_y,"Corrected: Random")
    # leg.AddEntry(h2_y_inv,"Corrected: Random (Inverse)")
    leg.AddEntry(h_jason_un_y,"Uncorrected: Non-Random")
    leg.AddEntry(h_jason_y,"Corrected: Non-Random")
    # leg.AddEntry(h2_y_inv,"Corrected: Non-Random (Inverse)")
    leg.Draw()

    c.SetLogy()
    c.SetGrid()

    c.Draw()
    c.Print("./images/PileupSpectrum_Energy_"+str(tlow).zfill(5)+"_"+str(thigh).zfill(5)+".png")
```

```python
c = r.TCanvas("c","c",1500,1400)
c.Divide(1,2)

h2.GetXaxis().UnZoom()
h_jason.GetXaxis().UnZoom()

h2.GetYaxis().SetRangeUser(3500,5900)
h2_x = h2.Project3D("x")
h2_x.SetTitle("Time Spectrum of Clusters Between 3500 and 5900 MeV (Pileup); Time [#mus]; N")

h2_x.GetXaxis().SetRangeUser(10,180)

h_jason.GetYaxis().SetRangeUser(3500,5900)
h_jason_x = h_jason.Project3D("x")
h_jason_x.GetXaxis().SetRangeUser(10,180)

# c.SetLogy()

for i in range(2):
    c.cd(i+1)
    h2_x.Draw()
    h_jason_x.Draw("same")

    r.gPad.SetGrid()
    
leg = r.TLegend(0.6,0.6,0.9,0.9)
leg.AddEntry(h2_x,"Corrected: Random")
leg.AddEntry(h_jason_x,"Corrected: Non-Random")
leg.Draw()

r.gPad.SetLogy()

c.Draw()
```
