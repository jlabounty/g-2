```python
for calo in range(1,25):
    print("Starting calo", calo)
    outfile = './outputTemplates/beamTemplatesByCalo/calotemplate'+str(calo)+'.root'
    c = r.TCanvas("c","c",15000,9000)
    c.Divide(9,6)
    msi = []
    ms2i = []
    for xtalNum in range(0,54):
        path = '/home/jlab/g-2/nearline/run3templates/templatesRun3/beamTemplates/calo'+str(calo)+'BeamTemplates/'
        fileName = 'template'+str(xtalNum)+'.root'

        file = path+fileName
        #print("Starting:", file)
        try:
            f = r.TFile(file)
            ms = f.Get("masterSpline").Clone("ms"+str(xtalNum))
        except:
            print("Unable to open file / spline", file)
            continue
        ms.SetTitle("calo"+str(calo)+"xtal"+str(xtalNum))
        
        out = r.TFile(outfile)
            
        ms2 = out.Get("masterSpline_xtal"+str(xtalNum)).Clone("ms2"+str(xtalNum))
        ms2.SetLineColor(2)
        
        msi.append(ms.Clone("ms"+str(xtalNum)))
        ms2i.append(ms2.Clone("ms2"+str(xtalNum)))
        
        out.Close()
        
    for i in range(len(msi)):
        c.cd(i+1)
        msi[i].Draw()
        ms2i[i].Draw("SAME")
        
    c.Draw()
    c.Print("./verification/test_calo"+str(calo).zfill(2)+".pdf")
    c.Print("./verification/test_calo"+str(calo).zfill(2)+".png")
```
