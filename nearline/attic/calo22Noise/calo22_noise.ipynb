```python
#f = r.TFile("./gm2nearline_hists_run32671_00121.root")
f = r.TFile("./results_run32671_histOnly.root")
```

```python
xtals = [(22,36),
         (22,37),
         (22,38),
         (21,37)
        ] #(calo, xtal)
```

```python
c = r.TCanvas()
colors = [2,3,4,6,7,8,9]
hists = []
leg = r.TLegend(0.6,0.7,0.9,0.9)
for i, (calo, xtal) in enumerate(xtals):
    hi = f.Get("nearlineHistTree/calo"+str(calo)+"/xtal"+str(xtal)+"Energy")
    hists.append( hi.Clone("calo"+str(calo)+"_xtal"+str(xtal)+"Energy") )
    color = colors[i]
    hists[i].SetLineColor(color)
    hists[i].SetFillColorAlpha(color, 0.2)
    hists[i].GetXaxis().SetRangeUser(0,2000)
    hists[i].GetYaxis().SetRangeUser(0.7,10000)
    if(i > 0):
        hists[i].Draw("SAME")
    else:
        hists[i].Draw("")
    leg.AddEntry(hists[i], "calo"+str(calo)+"_xtal"+str(xtal), "l")
leg.Draw()
c.SetLogy()
c.Draw()
```
