```python
from standardInclude import *
```

```python
f1 = r.TFile("./gm2nearline_ana_27415_00039_original.root")
f2 = r.TFile("./gm2nearline_ana.root")
```

```python
h1 = f1.Get("nearlineHistTree").Get("allCaloWiggle").Clone("h1")
h2 = f2.Get("nearlineHistTree").Get("allCaloWiggle").Clone("h2")
h2.SetLineColor(2)
```

```python
c = r.TCanvas()
h2.Draw()
h2.SetTitle("Wiggle Plot Before and After Change")
h1.Draw("Same")

leg = r.TLegend(0.6,0.8,0.9,0.9)
leg.AddEntry(h1,"Original Sync Pulse Corrector")
leg.AddEntry(h2,"New Sync Pulse Corrector")
leg.Draw()

c.SetLogy()
c.Draw()
```

```python
hdiff = h1.Clone("hdiff")
hdiff.Add(h2,-1)
```

```python
c = r.TCanvas()
hdiff.GetYaxis().SetRangeUser(-1,1)
hdiff.Draw()
c.Draw()
```
