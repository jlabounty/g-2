### Lets start with the energy of the positrons for the runs  24309  - 24352. During quad/scraping tests.

```python
f = r.TFile("./data/fullRuns_histsOnly/PositronClusters.root")
f.ls()
```

```python
c1 = f.Get("c1")
c1.Draw()
print([x for x in c1.GetListOfPrimitives()])
h = c1.GetPrimitive("h").Clone()
```

```python
c = r.TCanvas()
h.GetXaxis().SetRangeUser(0,950) 
h.SetTitle("Recon West Cluster Time vs. Energy; Time [#mus]; Energy [MeV]")
h.Draw("colz")
c.SetLogz()
c.Draw()
```

```python
c = r.TCanvas()
h.GetXaxis().SetRangeUser(800,850) 
h.Draw("colz")
c.SetLogz()
c.Draw()
```

```python
c = r.TCanvas()
h2 = h.ProjectionX().Clone()
h2.GetXaxis().SetRangeUser(800,825)
h2.Draw()
c.SetLogy()
c.Draw()
```

### Energies after 800 us by calo number

```python
f = r.TFile("./data/fullRuns_histsOnly/energies_calonum.root")
f.ls()
```

```python
c1 = f.Get("c1")
c1.Draw()
print([x for x in c1.GetListOfPrimitives()])
gr = c1.GetPrimitive("Graph")
```

```python
h = r.TH2D("h","Energy vs. CaloNum [t > 800 #mus]; Energy [MeV]; Calorimeter Number", 6000,0,6000,26,0,26)
for i in range(gr.GetN()):
    x = r.Double(-1)
    y = r.Double(-2)
    gr.GetPoint(i,x,y)
    h.Fill(x,y)
```

```python
c = r.TCanvas()
h.Draw("colz")
c.SetLogz()
c.Draw()
```

```python
c = r.TCanvas()
h.GetXaxis().SetRangeUser(0,800)
h.Draw("colz")
c.SetLogz()
c.Draw()

h.SetDirectory(0)
```
